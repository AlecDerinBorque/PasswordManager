````python
tmp_df = df[["Sales Volume", "price"]] #Creates a new DataFrame called tmp_df that only
                                      #contains two columns from the original df: "Sales Volume" and "price".
                                      #It's like making a copy with just those two columns so you don't mess with
                                      #your original data.

# tmp_df[:15].plot(kind='bar', ylabel="Normalized Value", xlabel="Item", title="Normalized Values bar chart")
# try this line above and you'll see why we use normilization

tmp_df['Sales Volume'] = (df['Sales Volume'] - df['Sales Volume'].min()) / \
                            (df['Sales Volume'].max() - df['Sales Volume'].min())
tmp_df['price'] = (df['price'] - df['price'].min()) / \
                            (df['price'].max() - df['price'].min())

#using min max normilization on sales volume and price features
#why use normalization? what ticked us off that we need to use it? why min max?

#we use min max for visualization. downside is that its very sensititve to outliers
#Example: If most sales are 100-200, but one freak sale is 10,000, min-max
#will make all the normal sales look tiny (close to 0).

#use z score for algoirhtms. z score can give negative numbers unlike min max which has range of (0,1)
#so (0,1) from min max is easier to visualize unlike possible negative nums from z score

tmp_df[:15].plot(kind='bar', ylabel="Normalized Value", xlabel="Item", title="Normalized Values bar chart")
````

# Removing Outliers

## Price

````python
df.describe()
````

````python
# before removing outlier prices
df.shape
````

````python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

# 2. Calculate IQR
IQR = Q3 - Q1
print(Q1, Q3)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
````

````python
df = df[(df["price"] >= lower_bound) & (df["price"] <= upper_bound)]
#removing outliers with this
````

````python
sns.pairplot(df, hue='season')
````

````python
# after removing outlier prices
df.shape
````

## Sales Volume (does nothing)

````python
df = df[(df["Sales Volume"] >= lower_bound) & (df["Sales Volume"] <= upper_bound)]
````

# Normalization

## Z-Score Normalization

````python
import numpy as np

normalized_df = df[["price", "Sales Volume"]]
mean = np.mean(normalized_df)
std = np.std(normalized_df)

df_normalized = (normalized_df - mean) / std
df_normalized
````

## Min-Max Normalization

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = df[["price", "Sales Volume"]]
fit = scaler.fit_transform(normalized_df)
normalized_df = pd.DataFrame(fit, columns=normalized_df.columns)
normalized_df
````
